```python
%%manim -qh Escena1_CreacionExciton

class Escena1_CreacionExciton(Scene):
    def construct(self):
        # 1. Crear la red cristalina
        atoms = VGroup()
        electrons = VGroup()
        for x in range(-4, 5):
            for y in range(-2, 3):
                atom = Circle(radius=0.15, color=RED, fill_opacity=0.5).move_to(RIGHT*x + UP*y*1.2)
                atoms.add(atom)
                elec = Dot(radius=0.08, color=BLUE).move_to(atom.get_center())
                electrons.add(elec)
        
        lattice = VGroup(atoms, electrons).center()
        self.play(Create(atoms), Create(electrons), run_time=1.5)
        
        # 2. Definir una función auxiliar para crear ondas
        def create_wave(start, end, color=YELLOW):
            p1, p2 = np.array(start), np.array(end)
            length = np.linalg.norm(p2 - p1)
            angle = np.arctan2(p2[1] - p1[1], p2[0] - p1[0])
            wave = ParametricFunction(
                lambda t: np.array([t, 0.2 * np.sin(2 * PI * t / 0.5), 0]),
                t_range=[0, length], color=color
            ).rotate(angle, about_point=ORIGIN).move_to((p1+p2)/2)
            return wave

        # Objetivo
        target_atom = atoms[22]
        target_elec = electrons[22]

        # Crear y animar el fotón
        photon = create_wave(LEFT*6 + UP*2, target_atom.get_center())
        label_photon = MathTex(r"h\nu \ge E_g", color=YELLOW, font_size=24).next_to(photon, UP)

        self.play(Create(photon), Write(label_photon), run_time=1)
        self.play(FadeOut(photon), FadeOut(label_photon)) # Impacto

        # 3. Excitación
        hole = Dot(radius=0.12, color=RED_E).move_to(target_atom.get_center())
        label_hole = Text("Hueco (+)", font_size=20, color=RED_E).next_to(hole, DOWN)
        excited_electron = target_elec.copy().set_color(BLUE_E)
        
        self.play(
            FadeOut(target_elec),
            FadeIn(hole), Write(label_hole),
            excited_electron.animate.shift(UP*1.5 + RIGHT*0.5)
        )

        # 4. Formación del Excitón (Enlace)
        coulomb_force = always_redraw(lambda: DashedLine(
            start=hole.get_center(), end=excited_electron.get_center(), color=PURPLE_B
        ))
        label_exciton = Text("Excitón", font_size=36, color=PURPLE).to_edge(UP)

        self.play(Create(coulomb_force), Write(label_exciton))

        # Baile orbital
        orbit_path = Circle(radius=1.5).move_to(hole.get_center())
        self.play(MoveAlongPath(excited_electron, orbit_path), run_time=3, rate_func=linear)
        self.wait(1)
```

```python
%%manim -qm Escena2_Reflectancia

class Escena2_Reflectancia(Scene):
    def construct(self):
        # Ejes
        axes = Axes(
            x_range=[0, 4, 1], y_range=[0, 1, 0.2],
            x_length=7, y_length=5,
            axis_config={"include_tip": True},
        ).to_edge(LEFT, buff=1)
        labels = axes.get_axis_labels(x_label="E (eV)", y_label="R")
        self.play(Create(axes), Write(labels))

        # Parámetros
        Eg, Eb = 3.0, 0.2
        E_exciton = Eg - Eb

        # Función de reflectancia simulada
        def r_func(x):
            base = 0.2 + 0.1 * x**2 if x < Eg else 0.5
            peak = 0.4 * np.exp(-((x - E_exciton)**2) / 0.01) # Pico gaussiano
            return base + peak

        # Curva
        graph = axes.plot(r_func, color=YELLOW, x_range=[0, 3.8])
        self.play(Create(graph), run_time=2)

        # Línea de Bandgap
        line_eg = DashedLine(axes.c2p(Eg, 0), axes.c2p(Eg, 1), color=GRAY)
        lbl_eg = MathTex("E_g", color=GRAY).next_to(line_eg, UP)
        self.play(Create(line_eg), Write(lbl_eg))

        # Escáner
        dot = Dot(color=RED)
        dot.move_to(axes.c2p(0, r_func(0)))
        
        # Visualización lateral
        exciton_viz = VGroup(Dot(color=RED), DashedLine(LEFT, RIGHT, color=PURPLE), Dot(color=BLUE)).scale(0.5).next_to(axes, RIGHT, buff=1)
        exciton_lbl = Text("Resonancia!", color=PURPLE, font_size=20).next_to(exciton_viz, UP)
        exciton_viz.set_opacity(0)
        exciton_lbl.set_opacity(0)
        self.add(exciton_viz, exciton_lbl)

        # Animación del recorrido
        self.play(
            MoveAlongPath(dot, graph),
            run_time=4,
            rate_func=linear
        )
        
        # Destacar el pico manualmente (simulado) al llegar cerca de 2.8 eV
        pico_point = axes.c2p(E_exciton, r_func(E_exciton))
        self.play(
            dot.animate.move_to(pico_point),
            exciton_viz.animate.set_opacity(1),
            exciton_lbl.animate.set_opacity(1),
            Flash(pico_point, color=PURPLE)
        )
        
        lbl_bind = MathTex("E < E_g", color=PURPLE).next_to(dot, UP)
        self.play(Write(lbl_bind))
        self.wait(2)
```

```python
%%manim -qh Escena3_Raman

class Escena3_Raman(Scene):
    def construct(self):
        title = Text("Efecto Raman (Stokes)", font_size=36).to_edge(UP)
        self.play(Write(title))

        # Cristal (Átomos y resortes)
        a1 = Circle(0.3, color=BLUE, fill_opacity=1).shift(LEFT*1.5)
        a2 = Circle(0.3, color=BLUE, fill_opacity=1)
        a3 = Circle(0.3, color=BLUE, fill_opacity=1).shift(RIGHT*1.5)
        atoms = VGroup(a1, a2, a3)
        
        # Resortes visuales
        line1 = always_redraw(lambda: Line(a1.get_right(), a2.get_left(), color=GRAY))
        line2 = always_redraw(lambda: Line(a2.get_right(), a3.get_left(), color=GRAY))
        
        crystal = VGroup(atoms, line1, line2).shift(DOWN)
        self.play(FadeIn(crystal))

        # 1. Fotón Incidente (Verde)
        # Usamos una función simple para la onda
        def get_wave(start, end, col, wl=0.5):
            p1, p2 = np.array(start), np.array(end)
            dist = np.linalg.norm(p2-p1)
            ang = np.arctan2(p2[1]-p1[1], p2[0]-p1[0])
            return ParametricFunction(
                lambda t: np.array([t, 0.2*np.sin(2*PI*t/wl), 0]),
                t_range=[0, dist], color=col
            ).rotate(ang, about_point=ORIGIN).move_to((p1+p2)/2)

        wave_in = get_wave(LEFT*5+UP*2, a2.get_center(), GREEN, 0.5)
        lbl_in = MathTex(r"\omega_{in}", color=GREEN).next_to(wave_in, UP)

        self.play(Create(wave_in), Write(lbl_in))
        self.play(FadeOut(wave_in), FadeOut(lbl_in), Flash(a2, color=WHITE))

        # 2. Vibración (Fonón creado)
        # Animación de oscilación
        self.play(
            a2.animate.shift(RIGHT*0.3),
            a1.animate.shift(LEFT*0.1),
            a3.animate.shift(LEFT*0.1),
            rate_func=wiggle, run_time=1.5
        )
        lbl_phonon = Text("Fonón emitido", font_size=20, color=GRAY).next_to(crystal, DOWN)
        self.play(Write(lbl_phonon))

        # 3. Fotón Saliente (Rojo - Menos energía)
        # Longitud de onda más larga (0.8)
        wave_out = get_wave(a2.get_center(), RIGHT*5+UP*2, RED, 0.8)
        lbl_out = MathTex(r"\omega_{out} < \omega_{in}", color=RED).next_to(wave_out, UP)

        self.play(Create(wave_out), Write(lbl_out))
        self.wait(2)
```

```python

from manim import *
import numpy as np

# Configuración básica
config.media_width = "75%"
config.verbosity = "WARNING"

# --- Funciones auxiliares para estas escenas ---
def crear_red_bonita(filas=4, columnas=6, espaciado=0.7):
    # Crea una red de puntitos brillantes
    atoms = VGroup()
    for i in range(columnas):
        for j in range(filas):
            # Usamos un Dot con un pequeño brillo alrededor
            core = Dot(radius=0.08, color=BLUE_C)
            glow = Dot(radius=0.15, color=BLUE_E, fill_opacity=0.3)
            atom = VGroup(glow, core).move_to(
                RIGHT * (i - columnas / 2 + 0.5) * espaciado +
                UP * (j - filas / 2 + 0.5) * espaciado
            )
            atoms.add(atom)
    return atoms.center()

# Función mejorada para ondas que entran y salen
def animar_onda(scene, start, end, color, run_time=1.5, label_tex=None, label_pos=UP):
    p1, p2 = np.array(start), np.array(end)
    dist = np.linalg.norm(p2-p1)
    angle = np.arctan2(p2[1]-p1[1], p2[0]-p1[0])
    
    # Usamos un objeto Wave que se mueve
    wave = ParametricFunction(
        lambda t: np.array([t, 0.15 * np.sin(3 * PI * t), 0]),
        t_range=[0, 1], color=color
    ).stretch_to_fit_width(dist).rotate(angle, about_point=ORIGIN).move_to((p1+p2)/2)
    
    # Máscara para que parezca que viaja
    mask = Rectangle(width=dist+1, height=1).align_to(p1, LEFT).rotate(angle, about_point=p1)
    
    wave_anim = Create(wave, run_time=run_time)
    
    if label_tex:
        label = Tex(label_tex, color=color, font_size=24).next_to(wave, label_pos)
        scene.play(wave_anim, Write(label))
        return VGroup(wave, label)
    else:
        scene.play(wave_anim)
        return wave
```

```python
%%manim -qh Intro_MateriaCondensada

class Intro_MateriaCondensada(Scene):
    def construct(self):
        # 1. El Caos (Física Atómica / Gas)
        title_chaos = Tex("Física Atómica (Individuos)").to_edge(UP)
        
        # Crear átomos en posiciones aleatorias
        num_atoms = 30
        random_atoms = VGroup()
        for _ in range(num_atoms):
            pos = np.array([np.random.uniform(-6, 6), np.random.uniform(-3, 3), 0])
            atom = Dot(radius=0.08, color=RED_C).move_to(pos)
            random_atoms.add(atom)
            
        self.play(Write(title_chaos), LaggedStart(*[GrowFromCenter(a) for a in random_atoms], lag_ratio=0.05))
        
        # Pequeño movimiento browniano para enfatizar el caos
        self.play(
            random_atoms.animate.shift(RIGHT*0.2),
            run_time=1, rate_func=wiggle
        )
        self.wait(0.5)
        
        # 2. La Condensación (El orden emergente)
        title_condensed = Tex("Materia Condensada (Colectivo)").to_edge(UP)
        
        # La red objetivo donde aterrizarán
        target_lattice = crear_red_bonita(filas=5, columnas=6, espaciado=0.8)
        
        # Animación de transición: Del caos al orden
        self.play(
            TransformMatchingShapes(title_chaos, title_condensed),
            Transform(random_atoms, target_lattice), # Los puntos aleatorios se transforman en la red
            run_time=3,
            path_arc=np.pi/2 # Hacemos que se muevan en arco, queda más bonito
        )
        
        # Enfatizar la unidad: toda la red vibra junta
        self.play(
            random_atoms.animate.scale(1.05),
            rate_func=there_and_back, run_time=1.5
        )
        
        subtitle = Tex("Propiedades Emergentes", color=YELLOW, font_size=30).next_to(random_atoms, DOWN)
        self.play(Write(subtitle))
        self.wait(2)
```

```python
%%manim -qh Intro_ElDialogo

class Intro_ElDialogo(Scene):
    def construct(self):
        # Título Conceptual
        main_title = Tex("El Diálogo Luz-Materia").to_edge(UP)
        self.play(Write(main_title))

        # 1. El Protagonista: El Cristal (La Orquesta)
        crystal = crear_red_bonita(filas=4, columnas=4, espaciado=1).shift(DOWN*0.5)
        label_crystal = Tex("Cristal", font_size=24, color=BLUE).next_to(crystal, DOWN)
        self.play(FadeIn(crystal), Write(label_crystal))
        self.wait(0.5)
        
        # Posiciones estratégicas para las ondas
        left_start = LEFT*6 + UP*0.5
        crystal_center = crystal.get_center() + UP*0.5
        right_end = RIGHT*6 + UP*0.5

        # --- PARTE 1: La Pregunta (Luz incidente) ---
        # Una onda amarilla llega al cristal
        wave_in = animar_onda(self, left_start, crystal_center, YELLOW, run_time=2, label_tex="Luz (Pregunta)", label_pos=UP)
        
        # Impacto visual: El cristal reacciona (se sacude entero)
        self.play(
            FadeOut(wave_in), # La onda desaparece al entrar
            crystal.animate.scale(1.1).set_color(YELLOW), # Se ilumina y crece
            rate_func=there_and_back, run_time=0.5
        )
        crystal.set_color(BLUE_C) # Vuelve a su color normal

        # --- PARTE 2: Las Tres Respuestas ---
        
        # RESPUESTA A: Reflectancia (Elástica)
        # Una onda igual rebota
        resp1_tex = Tex("1. Reflectancia (Rebote elástico)", color=YELLOW, font_size=30).to_edge(UP, buff=2)
        self.play(TransformMatchingShapes(main_title, resp1_tex))
        
        wave_out_refl = animar_onda(self, crystal_center, left_start + UP*1, YELLOW, run_time=2, label_tex="Misma Energía", label_pos=UP)
        self.wait(0.5)
        self.play(FadeOut(wave_out_refl))

        # RESPUESTA B: Excitón (Absorción/Resonancia Interna)
        resp2_tex = Tex("2. Excitón (Resonancia Interna)", color=PURPLE, font_size=30).to_edge(UP, buff=2)
        self.play(TransformMatchingShapes(resp1_tex, resp2_tex))
        
        # En lugar de salir una onda, aparece un brillo interno
        exciton_glow = Dot(radius=0.8, color=PURPLE).move_to(crystal.get_center()).set_opacity(0.5)
        exciton_label = Tex("Energía Absorbida", color=PURPLE, font_size=20).move_to(crystal.get_center())
        
        self.play(GrowFromCenter(exciton_glow), Write(exciton_label))
        # El cristal vibra suavemente mientras contiene la energía
        self.play(crystal.animate.shift(RIGHT*0.1), rate_func=wiggle, run_time=1)
        self.wait(1)
        self.play(FadeOut(exciton_glow), FadeOut(exciton_label))

        # RESPUESTA C: Raman (Inelástica)
        resp3_tex = Tex("3. Raman (Intercambio de Energía)", color=RED, font_size=30).to_edge(UP, buff=2)
        self.play(TransformMatchingShapes(resp2_tex, resp3_tex))
        
        # El cristal se queda vibrando (fonón)
        self.play(crystal.animate.shift(UP*0.1), rate_func=wiggle, run_time=2)
        
        # Mientras vibra, sale una onda ROJA (menos energía)
        # Nota: Usamos una función de onda diferente manual para hacerla más "larga" (menos energética)
        p1, p2 = np.array(crystal_center), np.array(right_end)
        dist = np.linalg.norm(p2-p1)
        wave_raman = ParametricFunction(
            lambda t: np.array([t, 0.2 * np.sin(1.5 * PI * t), 0]), # Frecuencia más baja (1.5 PI en vez de 3 PI)
            t_range=[0, dist], color=RED
        ).move_to((p1+p2)/2)
        
        label_raman = Tex("Menor Energía (Stokes)", color=RED, font_size=20).next_to(wave_raman, DOWN)
        
        self.play(Create(wave_raman), Write(label_raman), run_time=2)
        self.wait(2)
```

```python
%%manim -qh Escena_Acoplamiento

class Escena_Acoplamiento(Scene):
    def construct(self):
        title = Text("Acoplamiento Fuerte: El Polaritón").to_edge(UP)
        self.play(Write(title))

        # --- CONFIGURACIÓN DE LOS PÉNDULOS ---
        origin_left = UP*2 + LEFT*2
        origin_right = UP*2 + RIGHT*2
        length = 3.5
        
        # Péndulo 1: LUZ (Amarillo)
        rod1 = Line(origin_left, origin_left + DOWN*length, color=GRAY)
        bob1 = Circle(radius=0.4, color=YELLOW, fill_opacity=1).move_to(rod1.get_end())
        label1 = Text("Luz", font_size=24, color=YELLOW).next_to(bob1, DOWN)
        
        # Péndulo 2: MATERIA (Azul)
        rod2 = Line(origin_right, origin_right + DOWN*length, color=GRAY)
        bob2 = Circle(radius=0.4, color=BLUE, fill_opacity=1).move_to(rod2.get_end())
        label2 = Text("Materia", font_size=24, color=BLUE).next_to(bob2, DOWN)
        
        pendulum_group = VGroup(rod1, bob1, label1, rod2, bob2, label2)
        self.play(FadeIn(pendulum_group))
        
        # --- FASE 1: Separados (No hay interacción) ---
        # Movemos solo la luz, la materia se queda quieta
        self.play(
            bob1.animate.shift(LEFT*1),
            rod1.animate.rotate(0.2, about_point=origin_left),
            rate_func=there_and_back, run_time=1
        )
        self.play(Indicate(label2, color=RED, scale_factor=1.2)) # La materia ni se enteró
        
        # --- FASE 2: El Acoplamiento (Resorte) ---
        spring = always_redraw(lambda: DashedLine(
            bob1.get_center(), bob2.get_center(), color=GREEN, stroke_width=4
        ))
        coupling_text = Text("Interacción Fuerte", color=GREEN, font_size=30).move_to(UP*0.5)
        
        self.play(Create(spring), Write(coupling_text))
        
        # --- FASE 3: El Movimiento Conjunto (Modo Normal) ---
        # Ahora simulamos (simplificadamente) que se mueven juntos
        # Usamos una variable theta para oscilarlos
        
        tracker = ValueTracker(0)
        
        def update_pendulums(mob):
            t = tracker.get_value()
            # Ecuaciones simplificadas de acoplamiento (Modo simétrico)
            # Ambos se mueven, intercambiando energía
            theta1 = 0.3 * np.cos(3*t) 
            theta2 = 0.3 * np.cos(3*t - 0.5) # Pequeño desfase
            
            # Actualizar posiciones
            new_pos1 = origin_left + length * np.array([np.sin(theta1), -np.cos(theta1), 0])
            new_pos2 = origin_right + length * np.array([np.sin(theta2), -np.cos(theta2), 0])
            
            bob1.move_to(new_pos1)
            rod1.put_start_and_end_on(origin_left, new_pos1)
            label1.next_to(bob1, DOWN)
            
            bob2.move_to(new_pos2)
            rod2.put_start_and_end_on(origin_right, new_pos2)
            label2.next_to(bob2, DOWN)

        self.add(spring) # Asegurar que el resorte está
        
        self.play(
            UpdateFromFunc(pendulum_group, update_pendulums),
            tracker.animate.set_value(6), # Simular 6 segundos de tiempo
            run_time=6,
            rate_func=linear
        )

        # --- FASE 4: La Nueva Identidad ---
        final_box = RoundedRectangle(corner_radius=0.5, height=5, width=7, color=PURPLE)
        final_label = Text("Cuasipartícula: POLARITÓN", font_size=36, color=PURPLE).next_to(final_box, UP)
        
        self.play(Create(final_box), Write(final_label))
        self.wait(2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- CONFIGURACIÓN ESTILO "DARK MODE" ---
plt.style.use('dark_background')
colores = {
    'fondo': '#1e1e1e',     # Gris oscuro (tipo VS Code)
    'valencia': '#00d2ff',  # Cyan neón
    'conduccion': '#ff0055',# Magenta neón
    'texto': '#ffffff',
    'gap': '#222222'
}

def plot_diagrama_bloques():
    """Genera el gráfico simple de barras (Aislante vs Semiconductor)"""
    fig, ax = plt.subplots(figsize=(8, 6), facecolor=colores['fondo'])
    ax.set_facecolor(colores['fondo'])
    
    # Datos simulados
    ancho = 0.6
    x = [1, 2.5] # Posiciones
    
    # Banda de Valencia (Rectángulos llenos)
    ax.bar(x, [3, 3], width=ancho, bottom=0, color=colores['valencia'], alpha=0.8, label='B. Valencia (Lleno)')
    
    # Banda de Conducción (Rectángulos vacíos arriba)
    # Semiconductor (Gap pequeño)
    gap_semi = 1.1
    ax.bar(x[0], 2, width=ancho, bottom=3+gap_semi, color=colores['conduccion'], alpha=0.3, edgecolor=colores['conduccion'], lw=2)
    
    # Aislante (Gap grande)
    gap_aislante = 4.0
    ax.bar(x[1], 2, width=ancho, bottom=3+gap_aislante, color=colores['conduccion'], alpha=0.3, edgecolor=colores['conduccion'], lw=2)
    
    # Etiquetas y flechas
    # Flecha Gap Semiconductor
    ax.annotate(r'$E_g \approx 1 eV$', xy=(1, 3), xytext=(1, 3+gap_semi),
                arrowprops=dict(arrowstyle='<->', color='yellow', lw=2),
                ha='center', va='center', color='yellow', fontweight='bold', fontsize=12)
    
    # Flecha Gap Aislante
    ax.annotate(r'$E_g > 4 eV$', xy=(2.5, 3), xytext=(2.5, 3+gap_aislante),
                arrowprops=dict(arrowstyle='<->', color='yellow', lw=2),
                ha='center', va='center', color='yellow', fontweight='bold', fontsize=12)

    # Textos
    ax.text(1, 1.5, "Semiconductor\n(e.g., Silicio)", ha='center', color='black', fontweight='bold')
    ax.text(2.5, 1.5, "Aislante\n(e.g., Diamante)", ha='center', color='black', fontweight='bold')
    ax.text(1, 4.5, "Banda Conducción\n(Vacía)", ha='center', color=colores['conduccion'], fontsize=10)

    # Limpieza
    ax.set_ylim(0, 8)
    ax.set_xticks([])
    ax.set_ylabel("Energía (E)", fontsize=14, color='white')
    ax.set_title("Comparación de Bandgaps", fontsize=16, color='white', pad=20)
    
    # Eliminar bordes innecesarios
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    
    return fig

def plot_dispersion_Ek():
    """Genera el gráfico riguroso E vs k (Parábolas)"""
    fig, ax = plt.subplots(figsize=(8, 6), facecolor=colores['fondo'])
    ax.set_facecolor(colores['fondo'])
    
    k = np.linspace(-2, 2, 200)
    
    # Ecuaciones de Parábola (Masa efectiva)
    # E_v = -hbar^2 k^2 / 2mh
    E_valencia = -0.5 * k**2 
    
    # E_c = Eg + hbar^2 k^2 / 2me
    Eg = 1.5
    E_conduccion = Eg + 0.3 * k**2 
    
    # Plotear
    ax.plot(k, E_valencia, color=colores['valencia'], lw=3, label='Banda de Valencia')
    ax.plot(k, E_conduccion, color=colores['conduccion'], lw=3, label='Banda de Conducción')
    
    # Rellenar áreas (Sombreado bonito)
    ax.fill_between(k, E_valencia, -5, color=colores['valencia'], alpha=0.3)
    ax.fill_between(k, E_conduccion, 5, color=colores['conduccion'], alpha=0.1)
    
    # Flecha del Gap
    ax.annotate('', xy=(0, 0), xytext=(0, Eg),
                arrowprops=dict(arrowstyle='<->', color='yellow', lw=2, shrinkA=0, shrinkB=0))
    ax.text(0.1, Eg/2, r'$E_g$ (Bandgap)', color='yellow', fontsize=14, va='center')

    # Excitón (Visualización extra)
    # Dibujamos un nivel justo debajo de la banda de conducción
    E_exciton = Eg - 0.2
    ax.hlines(E_exciton, -0.4, 0.4, colors='white', linestyles='--', lw=1)
    ax.text(0.5, E_exciton, 'Nivel Excitónico', color='white', fontsize=10, va='center')

    # Etiquetas
    ax.set_xlabel(r'Momento Cristalino ($k$)', fontsize=14)
    ax.set_ylabel(r'Energía ($E$)', fontsize=14)
    ax.set_title(r'Relación de Dispersión $E(k)$', fontsize=16, pad=20)
    
    # Ejes centrados (Estilo libro de texto)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    
    ax.set_ylim(-2, 4)
    ax.set_xlim(-2, 2)
    
    # Quitar números del eje x (es conceptual)
    ax.set_xticks([])
    
    return fig

# --- GENERAR Y GUARDAR ---
fig1 = plot_diagrama_bloques()
fig1.savefig("bandgap_bloques.png", dpi=300, bbox_inches='tight', facecolor=colores['fondo'])

fig2 = plot_dispersion_Ek()
fig2.savefig("dispersion_Ek.png", dpi=300, bbox_inches='tight', facecolor=colores['fondo'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuración Dark Mode
plt.style.use('dark_background')

def plot_exciton_comparison():
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_aspect('equal')
    
    # 1. Dibujar la Red Cristalina (Átomos)
    # Creamos una grilla de puntos
    x = np.linspace(-10, 10, 21)
    y = np.linspace(-5, 5, 11)
    X, Y = np.meshgrid(x, y)
    
    # Átomos del cristal (Puntos grises)
    ax.scatter(X, Y, color='#444444', s=20, alpha=0.6, label='Átomos de la Red')

    # Centro (donde está el Hueco)
    ax.scatter(0, 0, color='red', s=100, zorder=10, label='Hueco (+)')

    # --- 2. EXCITÓN DE FRENKEL ---
    # Radio pequeño (~1 constante de red)
    # El electrón está en el mismo átomo o el vecino inmediato
    circle_frenkel = plt.Circle((-5, 0), 0.8, color='#ff0055', fill=False, lw=3)
    dot_frenkel = plt.Circle((-5, 0.8), 0.2, color='#00d2ff') # Electrón
    ax.add_patch(circle_frenkel)
    ax.add_patch(dot_frenkel)
    
    ax.text(-5, -2, "Excitón\nFrenkel", color='#ff0055', ha='center', fontweight='bold')
    ax.text(-5, -3, r"$r \approx a$", color='white', ha='center', fontsize=10)

    # --- 3. EXCITÓN DE WANNIER-MOTT ---
    # Radio grande (~5 constantes de red para este dibujo, en realidad es más)
    r_wannier = 4.5
    circle_wannier = plt.Circle((5, 0), r_wannier, color='#ffff00', fill=True, alpha=0.15, ls='--')
    border_wannier = plt.Circle((5, 0), r_wannier, color='#ffff00', fill=False, lw=2, ls='--')
    dot_wannier = plt.Circle((5 + r_wannier, 0), 0.2, color='#00d2ff') # Electrón orbitando lejos
    
    ax.add_patch(circle_wannier)
    ax.add_patch(border_wannier)
    ax.add_patch(dot_wannier)
    
    ax.text(5, -r_wannier - 1.5, "Excitón\nWannier-Mott", color='#ffff00', ha='center', fontweight='bold')
    ax.text(5, -r_wannier - 2.5, r"$r \gg a$", color='white', ha='center', fontsize=10)

    # Decoración
    ax.set_xlim(-11, 11)
    ax.set_ylim(-7, 7)
    ax.axis('off') # Quitar ejes para que parezca diagrama
    ax.set_title("Comparación de Radios Excitónicos", fontsize=16, color='white', pad=20)
    
    # Leyenda customizada
    ax.text(0, 6, "Constante de red ($a$) = Distancia entre puntos grises", ha='center', color='gray', fontsize=10)

    return fig

fig = plot_exciton_comparison()
plt.show()
fig.savefig("excitones_radios.png", dpi=300, facecolor='#1e1e1e')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Estilo Dark Mode
plt.style.use('dark_background')

def plot_niveles_excitonicos():
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Parámetros ficticios para visualizar
    Eg = 2.0       # Bandgap en eV
    Ry = 0.3       # Rydberg excitónico exagerado para que se vea bien (normalmente es ~0.01 eV)
    
    # Eje X (Solo para dibujar líneas)
    x_range = [-0.5, 0.5]
    
    # 1. Dibujar Bandas
    # Banda de Valencia (rectángulo abajo)
    ax.fill_between([-1, 1], -1, 0, color='#00d2ff', alpha=0.3)
    ax.text(0, -0.5, "Banda de Valencia\n(Llena)", ha='center', color='white', fontweight='bold')
    ax.axhline(0, color='#00d2ff', lw=2) # Borde superior BV
    
    # Banda de Conducción (rectángulo arriba)
    ax.fill_between([-1, 1], Eg, Eg+1, color='#ff0055', alpha=0.3)
    ax.text(0, Eg+0.5, "Banda de Conducción\n(Estados Libres)", ha='center', color='white', fontweight='bold')
    ax.axhline(Eg, color='#ff0055', lw=2) # Borde inferior BC
    
    # 2. Dibujar Niveles Excitónicos (Solución de Wannier)
    # E_n = Eg - Ry / n^2
    niveles = [1, 2, 3, 4] # n = 1, 2, 3...
    
    for n in niveles:
        E_n = Eg - Ry / (n**2)
        
        # Dibujar línea del nivel
        ax.hlines(E_n, -0.3, 0.3, colors='yellow', linestyles='-', lw=1.5)
        
        # Etiqueta
        label = f"n={n}"
        if n == 1:
            label += " (Fundamental)"
        ax.text(0.35, E_n, label, color='yellow', va='center', fontsize=9)
        
        # Flecha de enlace
        if n == 1:
             ax.annotate("", xy=(0, E_n), xytext=(0, Eg),
                        arrowprops=dict(arrowstyle="<->", color="white", lw=1))
             ax.text(-0.1, (E_n+Eg)/2, r"$E_b$ (Enlace)", color="white", fontsize=8, ha='right')

    # Continuum (n -> infinito)
    ax.text(0, Eg - 0.05, "n -> $\infty$ (Límite de ionización)", color='gray', fontsize=8, ha='center')

    # Decoración
    ax.set_xlim(-0.8, 0.8)
    ax.set_ylim(-0.2, Eg+0.8)
    ax.axis('off') # Sin ejes, solo diagrama conceptual
    
    ax.set_title(r"Soluciones de la Ecuación de Wannier: $E_n = E_g - R_y^*/n^2$", fontsize=14, color='white', pad=15)
    
    return fig

fig = plot_niveles_excitonicos()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Estilo Dark Mode
plt.style.use('dark_background')

def plot_frenkel_dispersion():
    fig, ax = plt.subplots(figsize=(10, 6), facecolor='#1e1e1e')
    ax.set_facecolor('#1e1e1e')
    
    # Rango de k (Zona de Brillouin) de -pi/a a +pi/a
    ka = np.linspace(-np.pi, np.pi, 300)
    
    # Parámetros
    E0 = 2.0  # Energía base (eV)
    J = -0.5  # Integral de transferencia (negativa para que el mínimo esté en k=0)
    
    # 1. Modelo Tight-Binding (Frenkel) -> COSENO
    # E(k) = E0 + 2J cos(ka)
    E_frenkel = E0 + 2 * J * np.cos(ka)
    
    # 2. Aproximación de Masa Efectiva (Wannier) -> PARÁBOLA
    # Cerca de k=0, cos(x) ~ 1 - x^2/2
    # E(k) ~ (E0 + 2J) - J(ka)^2 ... es una parábola
    # Ajustamos para que coincidan en el fondo
    E_wannier = (E0 + 2*J) + abs(J) * ka**2
    
    # Plotear
    ax.plot(ka, E_frenkel, color='#ff0055', lw=4, label=r'Frenkel (Tight-Binding): $E_0 + 2J\cos(ka)$')
    ax.plot(ka, E_wannier, color='#00d2ff', lw=2, ls='--', label=r'Wannier (Masa Efectiva): $\hbar^2 k^2 / 2m^*$')
    
    # Decoración
    ax.set_xlabel(r'Momento Cristalino ($ka$)', fontsize=14, color='white')
    ax.set_ylabel(r'Energía ($E$)', fontsize=14, color='white')
    ax.set_title("Relación de Dispersión: Frenkel vs Wannier", fontsize=16, color='white', pad=20)
    
    # Límites
    ax.set_ylim(0.5, 3.5)
    ax.set_xlim(-np.pi, np.pi)
    
    # Ticks bonitos para pi
    ax.set_xticks([-np.pi, 0, np.pi])
    ax.set_xticklabels([r'$-\pi$', '0', r'$+\pi$'])
    
    ax.legend(loc='upper center', fontsize=12)
    ax.grid(True, color='#333333', linestyle=':', alpha=0.6)
    
    # Anotación explicativa
    ax.text(0, 1.2, "El excitón se mueve\ncomo una onda", ha='center', color='white', fontsize=10)
    ax.annotate("", xy=(np.pi, E_frenkel[-1]), xytext=(np.pi, E0),
                arrowprops=dict(arrowstyle="<->", color="yellow"))
    ax.text(2.5, (E_frenkel[-1]+E0)/2, r"$2J$ (Ancho de Banda)", color="yellow", rotation=90, va='center')

    return fig

fig = plot_frenkel_dispersion()
plt.show()
fig.savefig("rela_disper.png", dpi=300, facecolor='#1e1e1e')
```

```python
%%manim -qh Escena_Polarizacion

class Escena_Polarizacion(Scene):
    def construct(self):
        title = Text("Origen Microscópico de la Reflexión").to_edge(UP)
        self.play(Write(title))

        # 1. Crear una hilera de átomos (La Superficie)
        atoms = VGroup()
        for i in range(5):
            # Núcleo (Rojo)
            core = Dot(color=RED, radius=0.15)
            # Nube electrónica (Azul - Anillo para ver el centro)
            cloud = Circle(color=BLUE, fill_opacity=0.5, radius=0.3).move_to(core)
            atom = VGroup(core, cloud).move_to(UP*0 + RIGHT*(i-2)*1.5)
            atoms.add(atom)
        
        surface_label = Text("Superficie del Cristal", font_size=24, color=GRAY).next_to(atoms, DOWN, buff=1)
        self.play(Create(atoms), Write(surface_label))

        # 2. Llega el Campo Eléctrico (Onda)
        # Una flecha que sube y baja representando E
        E_vector = Vector(UP, color=YELLOW).move_to(LEFT*4)
        E_label = MathTex(r"\vec{E}_{luz}", color=YELLOW).next_to(E_vector, LEFT)
        
        self.play(GrowArrow(E_vector), Write(E_label))

        # 3. Animación de Oscilación (Polarización)
        # Usamos un ValueTracker para simular el tiempo
        t = ValueTracker(0)

        # Función que mueve las nubes electrónicas en oposición al campo
        def update_atoms(mob):
            time = t.get_value()
            field_y = 0.5 * np.sin(3 * time) # El campo oscila
            
            # Actualizar flecha del campo
            new_vec = Vector(UP * field_y, color=YELLOW).move_to(LEFT*4)
            E_vector.become(new_vec)
            
            # Actualizar nubes (se mueven opuesto al campo por carga negativa)
            for atom in mob:
                core = atom[0] # Núcleo quieto
                cloud = atom[1] # Nube se mueve
                cloud.move_to(core.get_center() + DOWN * field_y * 0.5)
                
                # Efecto visual: Cuando la separación es máx, brilla
                opacity = 0.5 + 0.5 * abs(field_y)
                cloud.set_fill(opacity=opacity)

        self.add(atoms, E_vector)
        self.play(
            UpdateFromFunc(atoms, update_atoms),
            t.animate.set_value(2 * PI * 2), # 2 ciclos completos
            run_time=4,
            rate_func=linear
        )

        # 4. La Re-emisión (Ondas saliendo de los dipolos)
        # Congelamos la oscilación para explicar
        explanation = Text("Dipolos oscilantes = Antenas", color=BLUE_B, font_size=30).next_to(surface_label, DOWN)
        self.play(Write(explanation))
        
        # Emitir ondas desde cada átomo
        waves = VGroup()
        for atom in atoms:
            w = AnnularSector(inner_radius=0, outer_radius=2, angle=TAU, color=BLUE, fill_opacity=0).move_to(atom[0])
            waves.add(w)
            
        self.play(
            LaggedStart(
                *[Broadcast(atom[0], color=BLUE_E, run_time=2) for atom in atoms],
                lag_ratio=0.1
            )
        )
        
        final_text = Text("La suma de estas ondas es la 'Reflexión'", font_size=24, color=YELLOW).to_edge(DOWN)
        self.play(Write(final_text))
        self.wait(2)
```
